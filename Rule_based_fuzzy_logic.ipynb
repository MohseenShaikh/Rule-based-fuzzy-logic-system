1)   Represent the event probabilities in 2D array. Write a function in Python to calculate conditional probabilities P(A|B) and P(B|A).



```python
import numpy as np
```

```python
x = np.array([[20,30], [30, 20]])
```

```python
x
```

```python
import pandas as pd

df = pd.DataFrame(x,index=['Arts','Stem'], columns=['Male', 'Female'])
```

```python
df
```

```python
df.index
```

```python
def condition_prob(A):
  total = df[A].sum()
  joint_arts = df[A]['Arts'] 
  joint_stem = df[A]['Stem']
  arts_prob = joint_arts/total
  stem_prob = joint_stem/total
  print(f"Conditional probablity of Arts given {A} P(A|B): {arts_prob}")
  print(f"Conditional probablity of Stem given {A} P(A|B): {stem_prob}")
  return arts_prob, stem_prob

print("For given the male values")
arts_prob, stem_prob = condition_prob('Male')

print("For given the female values")
condition_prob('Female')
```

2)   Represent the rules as a list of antecedents and consequents (A-->B) along with their confidence values. Write a function in Python to calculate confidence value of the rules.



```python
data_dict = {'Antecedent': ['A', 'B'], 
             'Consequent': ['C', 'C'],
             'Certainity': [0.7,0.8]}
```

```python
rule_base = pd.DataFrame(data_dict, columns=['Antecedent', 'Consequent', 'Certainity'])
rule_base
```

```python
def calculate_confidence(ant, certainty, rule):
  df = rule[rule['Antecedent'].str.contains(ant)]
  action_certainty = df['Certainity'].values * certainty
  action_consequent = df['Consequent'].values
  return action_consequent[0], action_certainty[0]


consequent, certainty = calculate_confidence(ant='A',certainty=0.9, rule= rule_base)
print(f'Action consequent is {consequent} and Action certainty is {certainty}')
```
